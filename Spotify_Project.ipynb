```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
client_id = '=========Client ID =========='
client_secret = '============ Client Secret ============='

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
import pandas as pd
songs_list = []
offset = 0
while len(songs_list) < 1000:
    results = sp.search(q='tag:khmer OR tag:cambodia OR tag:"khmer song"', type='track', limit=50, offset=offset)
    for track in results['tracks']['items']:
        song_name = track['name']
        artist_name = track['artists'][0]['name']
        song_id = track['id']
        duration = track['duration_ms'] / 1000
        popularity_score = track['popularity']
        release_date = track['album']['release_date']
        song_type = track['album']['album_type']
        songs_list.append({'Song Name': song_name, 'Artist': artist_name, 'Duration': duration, 
                           'Popularity Score': popularity_score, 'Type': song_type, 'Release Date': release_date})
        if len(songs_list) >= 1000:
            break
    offset += 50

# Create a dataframe from the songs list
songs_df = pd.DataFrame(songs_list)
```
